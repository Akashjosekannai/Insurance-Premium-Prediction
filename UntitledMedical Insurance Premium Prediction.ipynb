### Checking Outliers

```python
# obtain the first quartile
Q1 = df.quantile(0.25)

# obtain the third quartile
Q3 = df.quantile(0.75)

# obtain the IQR
IQR = Q3 - Q1

# print the IQR
print(IQR)
```

```python
# filter out the outlier values
# ~ : selects all rows which do not satisfy the condition
# any() : returns whether any element is True over the columns
# axis : "1" indicates columns should be altered (0 for 'index')
df_iqr = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

## Spliting the data to Test & Train

```python
#spliting our feature and label data into train and test with scaled data
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(x_scaled,y,test_size =0.25,random_state = 8)
```

## KNN 

```python
#from sklearn.neighbors import KNeighborsRegressor
```

```python
mape_values = []

# Loop over a range of possible k values
for k in range(1, 21):
  # Create a knn regressor object with k as n_neighbors
  knn = KNeighborsRegressor(n_neighbors=k)
  # Fit the knn regressor on the training data
  knn.fit(X_train_s, y_train_s)
  # Make predictions on the testing data
  y_pred = knn.predict(X_test_s)
  # Calculate the mape for this k value
  mape_k = mape(y_test_s, y_pred)
  # Append the mape value to the list or dictionary
  mape_values.append(mape_k)

# Plot the mape values against the k values
plt.plot(range(1, 21), mape_values)
plt.xlabel('k')
plt.ylabel('mape')
plt.show()

# Find the k value that minimizes the mape
best_k = np.argmin(mape_values) +1
print(f'The best value of k is {best_k}')
```

```python
# Create a knn regressor object with k as n_neighbors
knn = KNeighborsRegressor(n_neighbors=3)
# Fit the knn regressor on the training data
knn.fit(X_train_s, y_train_s)
# Make predictions on the testing data
y_pred_test = knn.predict(X_test_s)
y_pred_train =knn.predict(X_train_s)
# Calculate the mape for this k value
mape_train_knn =mape(y_train_s,y_pred_train)
mape_test_knn = mape(y_test_s, y_pred_test)
print('For Train MAPE : ',mape_train_knn)
print('For Test MAPE  : ',mape_test_knn)
```

# Decision Tree

```python
#instantiating decision tree regressor
#criterion is absolute error
#hyperparameters are max depth
dtr = DecisionTreeRegressor( random_state = 5)
```

```python
# Applying Grid Search CV
# Define the parameter grid
param_grid = {
    'max_depth': [3,6,9,12,15],  # maximum depth of the tree
    'min_samples_split': [2,5,10],  # minimum number of samples required to split an internal node
    'min_samples_leaf': [1,2,3,10]  # minimum number of samples required to be at a leaf node
}
```

```python
# Define the scoring metric (e.g.,mae)
scoring = 'neg_mean_absolute_error'
```

```python
# Perform grid search
grid_search = GridSearchCV(dtr, param_grid, scoring=scoring)
grid_search.fit(X, y)
```

```python
# Retrieve the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
#instantiating decision tree regressor
#criterion is absolute error
#hyperparameters are max depth
dtr = DecisionTreeRegressor(max_depth= 12, min_samples_leaf= 1, min_samples_split=5 ,random_state = 5)
```

```python
#fitting the model
model_dtr = dtr.fit(X_train,y_train)
```

```python
#predicting our train data using decision tree regressor model
preds_train_dtr = model_dtr.predict(X_train) 

 #calculating mape for our train data
train_MAPE_dtr = mape(y_train, preds_train_dtr)

#printing train mape
print(train_MAPE_dtr) 
```

```python
#predicting our test data using decision tree regressor model
preds_test_dtr = model_dtr.predict(X_test) 

#calculating mape for our test data
test_MAPE_dtr = mape(y_test, preds_test_dtr) 

#printing test mape
print(test_MAPE_dtr)
```

# Random Forest

```python
#instantiating Random Forest regressor
forest = RandomForestRegressor()
```

```python
# Applying Grid Search CV
# Define the parameter grid
param_grid = {
    'max_depth': [3,6,9,12,15],  # maximum depth of the tree
    'min_samples_split': [2,5,10],  # minimum number of samples required to split an internal node
    'min_samples_leaf': [1,2,3,10]  # minimum number of samples required to be at a leaf node
}
```

```python
# Define the scoring metric (e.g.,mae)
scoring = 'neg_mean_absolute_error'
```

```python
# Perform grid search
grid_search = GridSearchCV(forest,param_grid, scoring=scoring)
grid_search.fit(X, y)
```

```python
# Retrieve the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
def random_forest_regression(X, y):
    # Train the random forest model.
    forest = RandomForestRegressor(max_depth=15, min_samples_leaf= 1, min_samples_split= 5,random_state=5)
    forest.fit(X_train, y_train)

    # Predict the test data.
    y_pred = forest.predict(X_test)

    # Calculate the train and test MAPE.
    train_mape = mean_absolute_percentage_error(y_train, forest.predict(X_train))
    test_mape = mean_absolute_percentage_error(y_test, y_pred)

    return train_mape, test_mape


 # Train and evaluate the random forest model.
train_mape, test_mape = random_forest_regression(X, y)

print("The train MAPE is:", train_mape)
print("The test MAPE is:", test_mape)
```

The Test MAPE for linear Regression Model is 3.80 %
